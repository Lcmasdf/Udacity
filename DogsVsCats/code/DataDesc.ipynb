```python
import os
import numpy as np
import pandas as pd
import cv2
from keras.preprocessing import image


def data_desc(path):
    size_array=[[],[]]
    max_size = [0,0]
    min_size = [99999999,99999999]
    for root, dirs, files in os.walk(path):
        for file in files:
            img = cv2.imread(path+'/'+file)
            height, width = img.shape[:2]
            size_array[0].append(height)
            size_array[1].append(width)
            if height*width > max_size[0]*max_size[1]:
                max_size = [height, width]
            if height*width < min_size[0]*min_size[1]:
                min_size = [height, width]
    return max_size,min_size,size_array
```

```python
import matplotlib.pyplot as plt

indexs = [1, 10, 100, 1000, 10000]
for i in range(5) :
    img = image.load_img('./train/cat.'+str(indexs[i])+'.jpg')
    plt.subplot(2,5,i+1)
    plt.imshow(img)
    plt.axis('off')
    img = image.load_img('./train/dog.'+str(indexs[i])+'.jpg')
    plt.subplot(2,5,6+i)
    plt.imshow(img)
    plt.axis('off')

plt.show()
    
```

```python
max_size, min_size, size_array = data_desc('./train')
plt.scatter(size_array[0], size_array[1])
print(max_size, min_size)
```

```python
plt.scatter(size_array[0], size_array[1])
plt.xlabel('height')
plt.ylabel('width')
plt.title('image size statistics')
```

```python
def img_show(image_names, row_num, column_num):
    images_index = zip(range(len(image_names)), image_names)
    for i, image_name in images_index:
        img = image.load_img('./train/'+image_name)
        plt.subplot(row_num, column_num, i+1)
        plt.imshow(img)
        plt.axis('off')
        plt.title(image_name)
    
    plt.show()

image_names = ['cat.6.jpg','cat.2337.jpg','cat.3216.jpg','dog.8671.jpg','dog.9517.jpg','dog.10237.jpg']
img_show(image_names, 2, 3)
```
